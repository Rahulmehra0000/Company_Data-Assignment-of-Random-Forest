```python
## Import Libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
company=pd.read_csv("/content/Company_Data.csv")
company.head()
```

```python
company['Sales_Efficiency'] = company.Sales.map(lambda x: 'High' if x>8 else 'Low')
```

```python
df= pd.get_dummies(company,columns=['ShelveLoc','Urban','US'])
```

```python
df['Sales_Efficiency'].value_counts()
```

```python
labelencoder = LabelEncoder()
df['Sales_Efficiency'] = labelencoder.fit_transform(df['Sales_Efficiency'])
```

```python
df['Sales_Efficiency'].value_counts()
## '0' = 'High'
## '1' = 'low'
## after lable encoder
```

# Outlier detiction by isolation forest

```python
from sklearn.ensemble import IsolationForest
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(df)
```

```python
# predictions
y_pred_outliers = clf.predict(df)
```

```python
y_pred_outliers 
```

```python
df['scores']=clf.decision_function(df)
```

```python
df
```

```python
df.columns.shape
```

```python
df['anomaly']=clf.predict(df.iloc[:,0:16])
## calculating anomaly socre
```

```python
df
```

```python
#Print the outlier data points
df[df['anomaly']==-1]
```

```python
Df=df.drop([5,75,159,207], axis=0,inplace=False)
## Here i had dropped multiple rows outlier from the data set
```

```python
Df
```

# Randomforest Model

```python
Feature_cols=['CompPrice','Income','Advertising','Population','Price','Education','ShelveLoc_Bad','ShelveLoc_Good','ShelveLoc_Medium','Urban_No','Urban_Yes','US_No','US_Yes']
```

```python
X = Df[Feature_cols]
Y = Df['Sales_Efficiency']
```

```python
kfold = KFold(n_splits=6, random_state=10)
model = RandomForestClassifier(n_estimators=10, max_features=3)
results = cross_val_score(model, X, Y, cv=kfold)
```

```python
results.mean()*100.0
## Accuracy
```
